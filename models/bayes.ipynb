```python
# using cuml's Naive Bayes classifier
# import cupy as cp
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, classification_report, hamming_loss, jaccard_score
from datetime import datetime
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
import scipy
import numpy as np
```

```python
def load_data():
    
    with np.load('/kaggle/input/smai-project/vectorised_data/X.npz', allow_pickle=False) as file:
#         print(file.__dict__)

        data = file['data']
        indices = file['indices']
        indptr = file['indptr']
        shape = file['shape']

        sparse_matrix = scipy.sparse.csr_matrix((data, indices, indptr), shape=shape, dtype=np.float32)

#     cupy_sparse_matrix = cp.sparse.csr_matrix(sparse_matrix)
    
    y = np.load('/kaggle/input/smai-project/vectorised_data/y.npy', allow_pickle=True)
#     print(y)

    X_train, X_test, y_train, y_test = train_test_split(sparse_matrix, y, test_size=0.2, random_state=48)
#     print(X_train)

#     print(type(X_train), type(X_test), type(y_train), type(y_test))
    
    return X_train, X_test, y_train, y_test
```

```python
# split data into train and test
X_train, X_test, y_train, y_test = load_data()

# create and fit model
model = OneVsRestClassifier(BernoulliNB(alpha=0.5))
model.fit(X_train, y_train)

# predict on test data
y_pred = model.predict(X_test)
# for i in range(5):
#     print(y_pred[i], y_test[i])

print(accuracy_score(y_test, y_pred))
print(1-hamming_loss(y_test, y_pred))
print(jaccard_score(y_test, y_pred, average='samples'))
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
# split data into train and test
# X_train, X_test, y_train, y_test = load_data()

# create and fit model
model = OneVsRestClassifier(MultinomialNB(alpha=0.5))
model.fit(X_train, y_train)

# predict on test data
y_pred = model.predict(X_test)

print(accuracy_score(y_test, y_pred))
print(1-hamming_loss(y_test, y_pred))
print(jaccard_score(y_test, y_pred, average='weighted'))
print(classification_report(y_test, y_pred))
```
